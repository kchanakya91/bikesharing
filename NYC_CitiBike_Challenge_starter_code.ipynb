```python
import pandas as pd
import datetime
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data.

file_dir = 'C:/Users/mrsme/Data Analysis Boot Camp/Challenges/bikesharing'
citibike_data = pd.read_csv(f'{file_dir}/201908-citibike-tripdata.csv', low_memory=False)
citibike_df = pd.DataFrame(citibike_data)
citibike_df.head()
```

```python
# 2. Check the datatypes of your columns.
citibike_df['tripduration_seconds'] = citibike_df['tripduration']

citibike_df.dtypes
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.
citibike_df['tripduration'] = pd.to_datetime(citibike_df['tripduration'], unit='s')
citibike_df.sample(5)
```

```python
# 4. Check the datatypes of your columns. 
citibike_df.dtypes
```

```python
# 5. Export the Dataframe as a new CSV file without the index.
# Create the output file (CSV).
output_data_file = "201908-citibike-tripdata_cleaned.csv"
# Export the city_data into a CSV.
citibike_df.to_csv(output_data_file, index=False)
```
